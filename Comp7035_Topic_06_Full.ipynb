Write a Python program to generate N times of die rolls where N should be updated and adjusted to get the desired result.  
Use Matplotlib to visualize the result and proof that the probability of each outcome (e.g. get the number '1') is approximately *0.16* - *0.17*

```python
import random
import matplotlib.pyplot as plt

rolls = 1000000

frequencies = [0,0,0,0,0,0]

for i in range(rolls):
        frequencies[random.randrange(1, 7) - 1] += 1

x = [1, 2, 3, 4, 5, 6]

prob = [x /rolls for x in frequencies]

plt.bar(x, prob)
plt.ylim(0.15, 0.17);
```

The easiest way to make a live animation in Matplotlib is to use one of the Animation classes.  
[matplotlib.animation](https://matplotlib.org/stable/api/animation_api.html)  

```python
import matplotlib.pyplot as plt
import matplotlib.animation
import numpy as np
from IPython.display import display, clear_output

t = np.linspace(0,2*np.pi)
x = np.sin(t)

fig, ax = plt.subplots()
l, = ax.plot([0,2*np.pi],[-1,1])

animate = lambda i: l.set_data(t[:i], x[:i])

for i in range(len(x)):
    animate(i)
    clear_output(wait=True)
    display(fig)

plt.show()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.animation
import numpy as np

t = np.linspace(0,2*np.pi)
x = np.sin(t)

fig, ax = plt.subplots()
l, = ax.plot([0,2*np.pi],[-1,1])

animate = lambda i: l.set_data(t[:i], x[:i])

ani = matplotlib.animation.FuncAnimation(fig, animate, frames=len(t))

from IPython.display import HTML
HTML(ani.to_jshtml())
```

Plot this figure the given alpha values:

$x = cos(\alpha)$

$y = sin(\alpha) + cos(\alpha)^{2/3}$

You can define $\alpha$ using the way we have learned for hart shape in our lecture notes.


```python
import numpy as np
from matplotlib import pyplot as plt

# Creating equally spaced 100 data in range 0 to 2*pi
alpha = np.linspace(0, 2 * np.pi, 100)

# Generating x and y data
x = np.cos(alpha)
y = np.sin(alpha) + np.cos(alpha)**(2/3)

# Plotting
plt.plot(x, y)
plt.title('Heart Shape')
plt.show()
```

### Animation

```python
import matplotlib.pyplot as plt
from matplotlib import animation
import numpy as np
from IPython.display import display, clear_output

fig = plt.figure() #Create a canvas to be painted
x = np.linspace(0, 10, 5)
y = x**2

ax = fig.subplots() #Create the subcanvas
l = ax.plot(x, y) #Define the data in x and y axix
l = l[0]

def animate(i):
  l.set_data(x[:i],y[:i])
  return l

for i in range(len(x)+1):
    animate(i)
    #clear_output(wait=True) #clear_output to clear the output of a cell.
    display(fig)
```

### Exercise for matplotlib


**Q1:** Plot the function $y = sin(x)$ for $x$ values ranging from $0$ to $2\pi$.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2*np.pi, 400)
y = np.sin(x)

plt.plot(x, y)
plt.title("y = sin(x)")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```



**Q2:** On the same graph, plot $y = sin(x)$ and $y = cos(x)$ for $x$ values ranging from 0 to $2\pi$. Add a legend to differentiate the two lines.

```python
y2 = np.cos(x)

plt.plot(x, y, label='sin(x)')
plt.plot(x, y2, label='cos(x)')
plt.legend()
plt.show()
```

```python
data = np.random.randn(1000)
plt.hist(data, bins=30)
plt.title("Histogram of Random Numbers from Normal Distribution")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

**Q5:** Generate two arrays of 100 random numbers each for x and y coordinates. Use another array of 100 random numbers as color values for each point. Display a scatter plot with a colorbar.

```python
x = np.random.rand(100)
y = np.random.rand(100)
colors = np.random.rand(100)

plt.scatter(x, y, c=colors, cmap='viridis')
plt.colorbar()
plt.title("Scatter Plot with Colorbar")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.show()
```


**Q6:** Given the following data, create a bar plot with error bars:

errors = [0.5, 1.2, 0.7]

```python
labels = ['A', 'B', 'C']
values = [10, 15, 7]
errors = [0.5, 1.2, 0.7]

plt.bar(labels, values, yerr=errors, capsize=5)
plt.title("Bar Plot with Error Bars")
plt.xlabel("Labels")
plt.ylabel("Values")
plt.show()
```

## Seaborn


**Question**: Load the `tips` dataset from `seaborn`. Use `sns.regplot()` to visualize the linear relationship between `total_bill` and `tip`.

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset('tips')
sns.regplot(x='total_bill', y='tip', data=tips)
plt.show()

```

**Question**: Using the tips dataset, create a regression plot of total_bill against tip, but create separate plots for lunch and dinner (time column).

```python
sns.lmplot(x='total_bill', y='tip', data=tips, col='time')
plt.show()
```

**Question:** Load the iris dataset from seaborn. Use sns.pairplot() to visualize pairwise relationships. Pay attention to the relationship between sepal_length and other numeric variables. Based on the plots, choose another variable apart from sepal_length that seems to have a linear relationship. Use sns.lmplot() to visualize a regression line.

```python
iris = sns.load_dataset('iris')
sns.pairplot(iris)
plt.show()

# Let's assume sepal_width seems to have a linear relationship with sepal_length
sns.lmplot(x='sepal_length', y='sepal_width', data=iris)
plt.show()
```

**Question:** In the tips dataset, visualize the effect of the day of the week on the relationship between total_bill and tip.

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='day')
plt.show()
```

**Question**: Sometimes, relationships are not linear but are better represented in a log scale. Using the tips dataset, transform the total_bill column to its logarithm and then plot a regression line against tip.


```python
import numpy as np

tips['log_total_bill'] = np.log(tips['total_bill'])
sns.regplot(x='log_total_bill', y='tip', data=tips)
plt.show()
```

**Question:** Using the tips dataset, explore the relationship between total_bill, tip, and size. Use sns.lmplot() to visualize a regression line of total_bill vs. tip, but this time, use the row or col parameter to separate the plots based on the size of the party.


```python
sns.lmplot(x='total_bill', y='tip', data=tips, col='size', col_wrap=3)
plt.show()
```

```python
# Extract the following Categorical variables and store as a new Pandas DataFrame.
houseCatData =houseData[['MSSubClass', 'Neighborhood', 'BldgType', 'OverallQual']]
houseCatData
```

```python
# b. Check the individual statistical description and visualize the distributions (catplot) of each of these variables.
print(houseCatData.describe())

for col in houseCatData.columns:
    sub = pd.DataFrame(houseCatData[col].value_counts()) # get the number of each catrgories in the single column
    sub = sub.reset_index() # extract the original index(category) as a independent column
    sns.catplot(data=sub,x="index",y=col,kind="bar") # show the distribution of the feature
```

```python
# c. Check the relationship amongst the variables using bi-variate heatmap of counts.
'''
I think the plot is meaningless in this way (as same as the link "What is heatmap of counts?" provided on Moodle), 
because there is no same classification between the two variables (it is different from the link example).
We can only get counts of each attributes of the single variable, and another one is null(0) at this time(the same value).
We cannot get meaningful info by this figure, or we can say we just got the counts of single variable.
'''
# show the bi-variate figures independently
# features = houseCatData.columns
# for i in range(len(features)-1):
#     for j in range(i+1,len(features)):
#         plt.figure()
#         sub = houseCatData[[features[i],features[j]]]
#         counts = sub.apply(sub.value_counts)
#         sns.heatmap(counts)

# show the heartmap of 4 variables
# counts=houseCatData.apply(houseCatData.value_counts)
# sns.heatmap(counts)


'''
I would like to try another way, to show how many entities that can satisfy the both condition.
For example, there're 2 categories A & B, A has 2 attributes a1,a2, B has b1, b2.
I would like to show the number of entities which can satisfy (a1,b1)/(a1,b2)/(a2,b1)/(a2,b2) at same time.
I think it should look like confusion matrix, but actually they represent different meanning.
So I can conclude that in generally, the situation of houses satisfy the (xx,xx) condition.
It may closer to the given figure on Moodle.
'''
features = houseCatData.columns
for i in range(len(features)-1):
    for j in range(i+1,len(features)):
        
        row = houseCatData[features[i]].unique() # get unique attributes of fisrt feature
        col = houseCatData[features[j]].unique() # get unique attributes of second feature
        
        counts=pd.DataFrame(columns=col,index=row) # build an empty df with 2 features' unique attr
        for r in row: # traversal all combined conditions
            for c in col:
                # count number of attr satisfied 2 conditions
                counts.loc[r,c] = ((houseCatData[features[i]]==r) & (houseCatData[features[j]]==c)).sum() 
        
        plt.figure() # show a new figure (otherwise will overlap)
        counts = counts.astype("float") # change type otherwise show error
        sns.heatmap(counts,cmap="YlGnBu") # show heatmap & change color (deepper color represents bigger number)
```

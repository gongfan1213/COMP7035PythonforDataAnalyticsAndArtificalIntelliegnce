### Problem 1: Creating a data structure of Pandas series
#### Given the index ['a', 'b', 'c', 'd', 'e'] and a randomized numpy array with 5 values,
##### Task 1.1: the first task is to create a Pandas series, and then print the result by setting the space of 50 between each index and value item 
##### Task 1.2: the second one is to output the result greater than 0.5. 

```python
import numpy as np
import pandas as pd

np.random.seed(1234)

# Task 1.1
value = np.random.rand(5)
index = ['a', 'b', 'c', 'd', 'e']

# Please write your code here
series = pd.Series(value, index)
print(series.to_string())

# Task 1.2
# Please write your code here
print("\nSubset of the above Data Series:")
print(series[series > 0.5].to_string())
```
